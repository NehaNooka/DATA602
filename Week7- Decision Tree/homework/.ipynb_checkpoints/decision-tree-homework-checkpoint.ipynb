```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.datasets import load_breast_cancer
df = load_breast_cancer()
dfX = df['data']
dfy = df['target']
dfNames = df['feature_names']
```

```python
from sklearn.model_selection import train_test_split
y = data['target']
X = data[[x for x in data.columns if x != 'target']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
X_train.drop(['target'],axis=1,inplace=True)
X_test.drop(['target'],axis=1,inplace=True)
```

## Using Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier(criterion='gini', random_state=1)
model1=tree_model.fit(X_train, y_train)

from sklearn import tree

plt.figure(figsize=(20,10))
tree.plot_tree(tree_model, feature_names=dfNames, fontsize=10)
plt.show()
```

```python
print(f'Test accuracy using Decision Tree is: {model1.score(X_test, y_test):.2f}')
```

## Using Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
logisticModel= LogisticRegression()
model2=logisticModel.fit(X_train, y_train)
```

```python
print(f'Test accuracy using Logistic Regression is : {model2.score(X_test, y_test):.2f}')
```

## Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=500)
model3=rnd_clf.fit(X_train, y_train)
```

```python
print(f'Test accuracy using Random Forest is : {model3.score(X_test, y_test):.2f}')
```

## Feature Importances

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

imps = list(zip(df.feature_names, rnd_clf.feature_importances_))

fi = pd.Series([x[1] for x in imps], index=[x[0] for x in imps])
fi = fi.sort_values(ascending=True)

fi.plot.barh()
plt.title('Feature Importances', loc='left')
plt.show()
```

## Random Forest performed best among all, so using Grid Search on it to perform better

```python
from sklearn.model_selection import GridSearchCV
params =  {'class_weight': [None, 'balanced'], 'C':[0.01, 0.1, 1, 10, 100],'max_iter':[1,100,1000,2000],
          'penalty':['l1', 'l2', 'elasticnet', 'none']}
# Instantiate the grid search model
grid_search = GridSearchCV(estimator=logisticModel,
                           param_grid=params,
                           cv = 4,
                           n_jobs=-1, verbose=1, scoring="accuracy")
grid_search.fit(X_train, y_train)
```

```python
print(f'Test accuracy using Grid Search with Logistic Reg : {grid_search.score(X_test, y_test):.2f}')
```

## There is not much improve in performance after using Grid Search. The accuracy increased from 93 to 94.
